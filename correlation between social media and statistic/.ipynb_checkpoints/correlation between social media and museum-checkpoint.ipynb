```python
ph_path = 'flickr_photo_infos_museum.csv' # produced from flickr_photo_in_museum.py
ph_df = pd.read_csv(ph_path,index_col=0)
ph_df.museum = ph_df.museum.apply(lambda x: x[:-1].replace(' ','_') if x[-1].isdigit() else x.replace(' ','_'))
def get_yr_mn_dy(x):
    dt = datetime.datetime.strptime(x,'%Y-%m-%d %H:%M:%S')
    return '%s_%02d_%02d' %(dt.year, int(dt.month),int(dt.day))
def get_yr_mn(x):
    dt = datetime.datetime.strptime(x,'%Y-%m-%d %H:%M:%S')
    return '%s_%02d' %(dt.year, int(dt.month))
ph_cnt_each_mu = {}
for mu, g in ph_df[['museum','owner','username','date_taken']].groupby('museum'):
    m_df = g.copy()
    m_df['ymd'] = m_df.date_taken.apply(lambda x: get_yr_mn_dy(x))
    m_df['ym'] = m_df.date_taken.apply(lambda x: get_yr_mn(x))
    m_df.drop_duplicates(['ymd','owner'],inplace=True)
    x = m_df.groupby('ym').count()['date_taken']
    x = x.reindex(months).fillna(0)
    ph_cnt_each_mu[mu]=list(x.values)
```

```python
df_flickr = pd.DataFrame(ph_cnt_each_mu)
df_flickr_norm = (df_flickr - df_flickr.min().min()) / (df_flickr.max().max() - df_flickr.min().min())
flickr_norm = df_flickr_norm.to_dict('list')
df_flickr_norm.shape
```

```python
months = ['2014_01','2014_02','2014_03','2014_04','2014_05','2014_06','2014_07','2014_08','2014_09','2014_10','2014_11','2014_12','2015_01','2015_02','2015_03','2015_04','2015_05','2015_06','2015_07','2015_08','2015_09','2015_10','2015_11','2015_12','2016_01','2016_02','2016_03','2016_04','2016_05','2016_06','2016_07','2016_08']
place_twitter_statistic = {}
min_month = {}
for place, g in twitter[['user','place','ymd','ym']].groupby('place'):
    pg = g.copy()
    pg.drop_duplicates(['ymd','user'],inplace=True)
    x = pg.groupby('ym').count()['ymd']
    min_month_idx = months.index(x.index.min())
    if min_month_idx<12:
        x = x.reindex(months[months.index(x.index.min()):]).fillna(0)
        x = x.reindex(months)
        place_twitter_statistic[place]=list(x.values)
        min_month[place]=min_month_idx
#     break
```

```python
place_twitter_statistic_df = pd.DataFrame(place_twitter_statistic)
pmax = place_twitter_statistic_df.max().max()
pmin = place_twitter_statistic_df.min().min()
place_twitter_statistic_df_norm = (place_twitter_statistic_df - pmin) / (pmax- pmin)
twitter_norm = place_twitter_statistic_df_norm.to_dict('list')
place_twitter_statistic_df_norm.shape
```
