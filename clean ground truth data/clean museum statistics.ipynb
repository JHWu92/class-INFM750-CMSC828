```python
data = u'..\\data\\museum_statistic_raw.csv'
clean_data = '..\\data\\museum_statistic_clean.csv'
import pandas as pd
import numpy as np
df = pd.read_csv(data)
df.head()
df = df.apply(lambda x: x.str.strip().replace('-',np.nan))
df.columns = ['mu']+list(df.columns[1:])
df
```

```python
melt_df = pd.melt(df, id_vars=['mu'],var_name='fy', value_name='cnt')
melt_df = melt_df[~melt_df.mu.str.contains('_FY_Total')]
melt_df['month'] = melt_df['mu'].apply(lambda x: x.rsplit('_',1)[1])
melt_df['mu'] =  melt_df['mu'].apply(lambda x: x.rsplit('_',1)[0])
```

```python
mn2int = {'April':4, 'May':5, 'June':6, 'July':7, 'August':8, 'September':9, 'October':10, 'November':11, 'December':12, 'January':1, 'February':2, 'March':3}
def get_yr_mn(x):
#     print x
    mn = mn2int[x.month]
    fy = x.fy[2:].split('_')
    yr = '20%s' % fy[0 if mn>3 else 1]
    return '%s_%02d' % (yr, mn)
melt_df['ym'] = melt_df.apply(get_yr_mn,axis=1)
melt_df.to_csv(clean_data)
```
