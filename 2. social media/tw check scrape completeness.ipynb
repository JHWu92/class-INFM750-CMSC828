```python
import pandas as pd
import glob 
def get_grp_paths(_dir):
    file_list = list(glob.glob(_dir+'*.*'))
    df_file = pd.DataFrame(file_list, columns=['path'])
    df_file['filename'] = df_file.apply(lambda x: os.path.basename(x.values[0]),axis=1)
    df_file['place'] = df_file.filename.apply(lambda x: x.rsplit('_',2)[0])
    for place, grp in df_file.groupby('place'):
        print place, grp.shape
        yield place, grp

def read_data(_dir):
    for place, grp in get_grp_paths(_dir):
        dfs = grp.path.apply(lambda x: pd.read_csv(x, header=None, sep='\t', encoding='utf-8')  if os.path.getsize(x)!=0 else pd.DataFrame())
        df_place = pd.concat(dfs.values,ignore_index=True).shape

read_data(sm_raw_dir)
```
